### Data preperation

Before extracting the data first we have to prepare the images and apply some filters for the classifier to make the best use of the features.
<br>
We concluded from our research, that demented brains have more black areas around the brain so the edges and the color difference must be clear  
<br>
In our case, **we used two filters and the third one is a combination of the two and compared them**

```python
def sobel_filter(image):
    img = cv2.imread(image)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobimg = sobel(new_img)
    return sobimg
```

```python
def my_filter(image):
    filimg = cv2.imread(image)
    ret, thresh = cv2.threshold(cv2.cvtColor(filimg,cv2.COLOR_BGR2GRAY),0,255,cv2.THRESH_OTSU)
    ret, markers = cv2.connectedComponents(thresh)
    marker_area = [np.sum(markers==m) for m in range(np.max(markers)) if m!=0]
    if marker_area == []:
        marker_area = [6]
    largest_component = np.argmax(marker_area)+1
    brain_mask = markers==largest_component
    brain_out = filimg.copy()
    brain_out[brain_mask==False] = (0,0,0)
    brain_out = cv2.cvtColor(brain_out, cv2.COLOR_BGR2GRAY)
    return brain_out
```

```python
def my_filterXsobel(image):
    filimg = cv2.imread(image)
    ret, thresh = cv2.threshold(cv2.cvtColor(filimg,cv2.COLOR_BGR2GRAY),0,255,cv2.THRESH_OTSU)
    ret, markers = cv2.connectedComponents(thresh)
    marker_area = [np.sum(markers==m) for m in range(np.max(markers)) if m!=0]
    if marker_area == []:
        marker_area = [6]
    largest_component = np.argmax(marker_area)+1
    brain_mask = markers==largest_component
    brain_out = filimg.copy()
    brain_out[brain_mask==False] = (0,0,0)
    brain_out = cv2.cvtColor(brain_out, cv2.COLOR_BGR2GRAY)
    sobimg = sobel(brain_out)
    return sobimg
```

```python
def plot_images(dir):
    imgdir = dir
    img1 = plt.imread(imgdir)
    img2 = sobel_filter(imgdir)
    img3 = my_filter(imgdir)
    img4 = my_filterXsobel(imgdir)
    titles = ['Original Image', 'sobel', 'black filter', 'both']
    images = [img1, img2, img3, img4]
    plt.figure(figsize=(15,15))
    for i in range(4):
        plt.subplot(1,4,i+1),plt.imshow(images[i],'gray')
        plt.title(titles[i])
        plt.xticks([]),plt.yticks([])
    plt.show()
```

    <li>KNN will take a shorter time at the time of training since it only places the neighbors when training, unlike the classification time it will take longer since it needs to calculate the distance for all neighbors</li>

```python
def test_model(mod, x, y):
    kf = KFold(n_splits=10)
    mae_train = []
    mae_test = []
    for train_index, test_index in kf.split(x):
        
        X_train, X_test = x.iloc[train_index], x.iloc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        model = mod
        model.fit(X_train, y_train)
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)
        mae_train.append(mean_absolute_error(y_train, y_train_pred))
        mae_test.append(mean_absolute_error(y_test, y_test_pred))

    folds = range(1, kf.get_n_splits() + 1)
    plt.plot(folds, mae_train, 'o-', color='green', label='train')
    plt.plot(folds, mae_test, 'o-', color='red', label='test')
    plt.legend()
    plt.grid()
    plt.xlabel('Number of fold')
    plt.ylabel('Mean Absolute Error')
    plt.show()
```

```python
xdf = pd.DataFrame(scaled_data)
test_model(svm_model, xdf, y)
```
